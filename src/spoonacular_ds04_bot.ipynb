```python
'''
Телеграм-бот CookingBot (@spoonacular_ds04_bot) - выводит по запросу рецепты блюд
Создан командой "Dream Team" в рамка проекта Data Scientist. Project 04 "REST-API" Школы 21 Сбербанка
При первом запуске бота нажать "Start" или ввести "/start"
Кнопок - 3
Ограничение - 150 рецептов в день
'''

from Recipes import *
import telebot
from telebot import types

# Ключ для запросов по API с сайта https://spoonacular.com
apikey = "some api key" 
# Токен Телеграм-бота
token = 'telegram api token'   # @spoonacular_ds04_bot

msg_start = 'I am here :)\nIf there are no buttons on the screen, click the Buttons icon...'
btn1_text = 'Random'
btn2_text = 'ByDish'
btn3_text = 'ByIngr'
msg2_bot = 'Please enter the name of the dish...'
msg3_bot = 'Please list the ingredients separated by commas...'

flag2_input_wait = 0
flag3_input_wait = 0

if __name__ == '__main__':
    # Создание экземпляра класса
    client = Recipes(apikey=apikey)

    # Инициализация бота
    bot = telebot.TeleBot(token, parse_mode=None)

    # Обработчик команды /start
    @bot.message_handler(commands = ["start"])
    def start(message, res=False) :
        markup = types.ReplyKeyboardMarkup(resize_keyboard=True)
        item1 = types.KeyboardButton(btn1_text)
        item2 = types.KeyboardButton(btn2_text)
        item3 = types.KeyboardButton(btn3_text)
        markup.add(item1, item2, item3)
        bot.send_message(message.chat.id, msg_start, reply_markup=markup)        
    
    # Обработчик сообщений от пользователя
    @bot.message_handler(content_types=["text"])
    def handle_text(message) :
        global flag2_input_wait
        global flag3_input_wait
            
        if message.text == btn1_text :
            flag2_input_wait = 0
            flag3_input_wait = 0
            answer=client.get_random_recipe()
        elif message.text == btn2_text :
            flag2_input_wait = 1
            flag3_input_wait = 0
            message.text == btn2_text
            answer = msg2_bot
        elif message.text == btn3_text :
            flag2_input_wait = 0
            flag3_input_wait = 1
            message.text == btn3_text
            answer = msg3_bot
        else :  
            if flag2_input_wait == 1:    
                answer=client.get_recipe(message.text)
            elif flag3_input_wait == 1:
                answer=client.get_recipe_by(message.text)
            else:
                answer= 'Please press one of the buttons...' 
        
        if (answer == None) | (answer == "error!"):
            answer = 'Sorry, the recipe {'+ message.text + '} was not found, try again...'+' :return '+answer
            
        # Отправка сообщения пользователю в чат
        bot.send_message(message.chat.id, answer)
    
    # Перевод бота в режим прослушивания команд
    bot.infinity_polling(none_stop=True, interval=0)  
```
